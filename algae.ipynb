```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./algae.csv')
```

```python
sns.lineplot(data=df, x='day', y = 'color_score', hue='condition', alpha = 0.4, linewidth = 8)
plt.legend(title='Condition', title_fontsize=16, 
           labels=['Control', 'Commerical Algaecide', 'Chlorine', 'Salt'], fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(range(1,6,1), fontsize=16)
plt.xlabel("Day", fontsize=16)
plt.ylabel("Color Score", fontsize=16)
plt.savefig('colorscore.png')
plt.show()
```

```python
df_salt = df[df.condition == 'salt']
df_chlorine = df[df.condition == 'chlorine']
df_commercial_algaecide = df[df.condition == 'commercial_algaecide']
df_control = df[df.condition == 'control']

dfs = [df_salt, df_chlorine, df_commercial_algaecide, df_control]
```
